```python
from manim import *
```

```python
%%manim -v WARNING --disable_caching -qh GraphThing

config.media_width = "50%"
config.media_embed = True

class GraphThing(Scene):
    def construct(self):
        ax = Axes(
            x_range=[0,10],
            y_range=[0,10],
            axis_config={"include_numbers": True},
        )

        labels = ax.get_axis_labels()

        self.wait()
        self.play([Create(ax), Create(labels)], run_time=3)
        self.wait()
```
