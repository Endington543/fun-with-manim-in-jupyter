```python
%%manim -v WARNING -pqh --progress_bar leave GraphThing

config.media_width = "50%"
config.media_embed = True


class GraphThing(Scene):
    def construct(self):
        ax = Axes(
            x_range=[0,20.01],
            y_range=[-10,10.01,2],
            axis_config={"include_numbers": True},
            tips=False,
        )

        labels = ax.get_axis_labels()
        t = MathTex(r"y = \sin(x)")

        t.next_to(ax.get_x_axis(), UP)

        sin_func = ax.plot(
            lambda x: np.sin(x),
            color=PINK
        )

        self.wait()
        self.play([Create(ax), Write(labels)], run_time=3)
        self.wait()
        self.play(Write(t))
        self.wait()
        self.play(t.animate.to_corner(UR))
        self.wait()
        self.play(Create(sin_func))
        self.wait()

        t2 = MathTex(r"y = 8\sin(x)").to_corner(UR)

        self.play(ReplacementTransform(t, t2))
        self.wait()

        sin_amp8 = ax.plot(
            lambda x: 8*np.sin(x),
            color=sin_func.get_color(),
        )
        
        self.play(ReplacementTransform(sin_func, sin_amp8), run_time=2)
        self.wait()

        t3 = MathTex(r"y = 8\sin(x) + 3\cos(x)").to_corner(UR)

        self.play(ReplacementTransform(t2, t3))
        self.wait(0.5)

        sin_comp_func = ax.plot(
            lambda x: 8*np.sin(x) + 3 * np.cos(x),
            color=GREEN,
        )

        self.play(
            TransformFromCopy(sin_amp8, sin_comp_func)
        )
        self.wait()

        t4 = MathTex(r"y = 8\sin(x) + 3\cos(x) * \frac{14}{7}\cos(x)").to_corner(UR)

        self.play(ReplacementTransform(t3, t4))
        self.wait(0.5)

        sin_comp_func1 = ax.plot(
            lambda x: 8*np.sin(x) + 3 * np.cos(x) * (14 / 7) * np.cos(x),
            color=BLUE,
        )

        self.play(TransformFromCopy(sin_comp_func, sin_comp_func1))
        self.wait()
```

```python
%%manim -v WARNING -qh --disable_caching Test

config.media_width = "60%"

class Test(Scene):
    def construct(self):
        t1 = MathTex("{{x}}^2", "=", "3{{x}}", "+", "32", font_size = 100, color=PINK)
        t2 = MathTex("{{x}}^2", "-", "3{{x}}", "-", "32", "=", "0", font_size=100, color=PINK)

        self.wait()
        self.play(Write(t1), run_time=2)
        self.wait()
        self.play(TransformMatchingTex(t1,t2), run_time=2)
        self.wait()
```
